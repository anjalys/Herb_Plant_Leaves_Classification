Currently the system they use to classify the trees which they import in a batch is quite manual. A labourer from his experience decides the leaf type and subtype of plant family. They have asked us to automate this process and remove any manual intervention from this process.

```python
from sklearn.preprocessing import LabelEncoder

#function to encode the label
def encode(label):
  le = LabelEncoder()
  encoded_label = le.fit_transform(label)

  return encoded_label
```

```python
df_train['species'] = encode(df_train['species'])
```

```python
df_train.head()
```

```python
df_train.drop('id', axis=1, inplace=True)
```

```python
df_train.head()
```

```python
X = df_train.drop('species', axis=1)
y = df_train['species']
```

```python
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import accuracy_score
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
#classifier object 
svc = SVC()

#fit the model
svc.fit(X_train, y_train)

pred_svc = svc.predict(X_test)

accuracy_score(y_test, pred_svc)
```

```python
#initialize decision tree classifier object
tree = DecisionTreeClassifier()

tree.fit(X_train, y_train)

pred_tree = tree.predict(X_test)

accuracy_score(y_test, pred_tree)
```

```python
#initialize Naive Bayes
naive = GaussianNB()

pred_naive = naive.fit(X_train, y_train).predict(X_test)

accuracy_score(y_test, pred_naive)
```

```python
forest = RandomForestClassifier()

pred_forest = forest.fit(X_train, y_train).predict(X_test)

accuracy_score(y_test, pred_forest)
```
